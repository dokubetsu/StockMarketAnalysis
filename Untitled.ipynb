```python
train_data = pd.DataFrame(df['Close'][0:int(len(df)*0.70)])
test_data = pd.DataFrame(df['Close'][int(len(df)*0.70):int(len(df))])
```

```python
train_data
```

```python
test_data
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
train_data_array = scaler.fit_transform(train_data)
train_data_array
```

```python
test_data_array = scaler.fit_transform(test_data)
test_data_array
```

```python
x_train = []
y_train = []

for i in range(60, train_data_array.shape[0]):
    x_train.append(train_data_array[i-60:i])
    y_train.append(train_data_array[i, 0])     
```

```python
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
model = Sequential()
model.add(LSTM(units = 50,  activation ='relu', return_sequences =True, input_shape=(x_train.shape[1],1)))
model.add(Dropout(0.2))

model.add(LSTM(units = 75,  activation ='relu', return_sequences =True, input_shape=(x_train.shape[1],1)))
model.add(Dropout(0.3))

model.add(LSTM(units = 100,  activation ='relu', return_sequences =True, input_shape=(x_train.shape[1],1)))
model.add(Dropout(0.4))

model.add(LSTM(units = 125,  activation ='relu'))
model.add(Dropout(0.5))
          
model.add(Dense(units=1))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
model.fit(x_train,y_train,epochs=50)
```

```python
model.save('keras_model.h5')
```

```python
days30 = train_data.tail(60)
final_df = days30._append(test_data, ignore_index=True)
final_df
```

```python
input_data = scaler.fit_transform(final_df)
```

```python
x_test = []
y_test = []

for i in range(60,input_data.shape[0]):
    x_test.append(input_data[i-60: i])
    y_test.append(input_data[i, 0])
```

```python
x_test, y_test = np.array(x_test), np.array(y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
scaler = scaler.scale_
```

```python
scale_factor = 1/scaler
y_pred = y_pred*scale_factor
y_test = y_test*scale_factor
```

```python
plt.figure(figsize = (20,10))
plt.plot(y_test,'b',label='Original_Value')
plt.plot(y_pred,'r',label='Predicted_value')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show
```
